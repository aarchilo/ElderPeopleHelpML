```python
import pandas as pd
import seaborn as sns
from sklearn.impute import SimpleImputer


df=pd.read_csv('dirty_v3_path.csv')
df.head()
#finding data which is null
df.dropna(subset=['Medical Condition'], inplace=True)
print(df.isnull().sum())

#normal distributed
sns.kdeplot(df['Age'], shade=True, color='skyblue')
sns.kdeplot(df['Glucose'], shade=True, color='red')
sns.kdeplot(df['Blood Pressure'], shade=True, color='yellow')
```

```python
mean_imputer = SimpleImputer(strategy='mean')
median_imputer = SimpleImputer(strategy='median')

# Apply mean imputation to numeric columns
df[['Age', 'Blood Pressure']] = mean_imputer.fit_transform(df[['Age', 'Blood Pressure']])
#its not normally distributed so 
df[['Glucose']]=median_imputer.fit_transform(df[['Glucose']])
```

```python
#changed the null values
df.isnull().sum()
#thing not affecting over relation
df.drop(columns=['noise_col','random_notes','Gender'],inplace=True)
#data split in output and input format
y=df['Medical Condition']
df.drop(columns=['Medical Condition'],inplace=True)
x=df.iloc[:,:]
print(df.isnull().sum())
sns.kdeplot(df['Age'], shade=True, color='skyblue')
sns.kdeplot(df['Glucose'], shade=True, color='red')
sns.kdeplot(df['Blood Pressure'], shade=True, color='yellow')
```
